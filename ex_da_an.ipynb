```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_data = pd.read_csv('Data/train.csv', index_col='id')
test_data = pd.read_csv('Data/test.csv', index_col='id')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 10))  # Increased figsize
axes = axes.flatten()
segmentation_data = train_data[['Sex', 'Chest pain type', 'EKG results', 'Exercise angina', 'Slope of ST', 'Number of vessels fluro', 'Thallium']]
for ax, col in zip(axes, segmentation_data.columns):
    value_counts = segmentation_data[col].value_counts()
    ax.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(col, fontsize=10)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(20, 5))
cont_data = train_data[['Age', 'BP', 'Cholesterol', 'Max HR', 'ST depression']]
# Plot histograms
for ax, col in zip(axes, cont_data.columns):
    sns.histplot(train_data[col], ax=ax, kde=True)
    ax.tick_params(labelsize=10)
plt.tight_layout()
plt.show() 
```
